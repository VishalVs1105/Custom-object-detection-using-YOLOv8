```python
from ultralytics import YOLO
import cvzone
import cv2
import math
from pygame import mixer
from pync import Notifier
#from twilio.rest import Client
```

```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
```

```python
cap = cv2.VideoCapture(1)
model = YOLO('best1.pt')
```

```python
classnames = ['pistol','knife']
```

```python
mixer.init()
mixer.music.load('alarm.wav')
mixer.music.set_volume(0.9)
```

```python
smtp_server = 'smtp.gmail.com'  # Replace with your SMTP server address
smtp_port = 587  # This is the typical SMTP port, but it may vary
sender_email = 'abc@gmail.com'
sender_password = 'passs'
```

```python
server = smtplib.SMTP(smtp_server, smtp_port)
server.starttls()  # Use if your SMTP server requires a secure connection
server.login(sender_email, sender_password)
subject = 'Danger!!!'
message = 'Alert!!!! detected!!!'
email_text = MIMEMultipart()
email_text['From'] = sender_email
email_text['To'] = 'vsvishal2013@gmail.com'  # Replace with the recipient's email address
email_text['Subject'] = subject
email_text.attach(MIMEText(message, 'plain'))
```

```python
while True:
    ret,frame = cap.read()
    frame = cv2.resize(frame,(640,480))
    result = model(frame,stream=True)
    confi = 0

    # Getting bbox,confidence and class names informations to work with
    for info in result:
        boxes = info.boxes
        for box in boxes:
            confidence = box.conf[0]
            confidence = math.ceil(confidence * 100)
            confi = confidence
            Class = int(box.cls[0])
            if confidence > 50:
                title = 'Notification'
                message = 'gun detected.'
                Notifier.notify(message, title=title, timeout=10) #laptop
                mixer.music.play() #music
                '''client = Client(twilio_account_sid, twilio_auth_token)
                sms_message = client.messages.create(
                body='Alert!!! Gun Detected!!!!',
                from_=twilio_phone_number,
                to=recipient_phone_number
                )# sms'''
                recipient_email = 'richieddy4@gmail.com'  # Replace with the recipient's email address
                server.sendmail(sender_email, recipient_email, email_text.as_string())
                x1,y1,x2,y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1),int(y1),int(x2),int(y2)
                cv2.rectangle(frame,(x1,y1),(x2,y2),(0,0,255),5)
                cvzone.putTextRect(frame, f'Pistol {confidence}%', [x1 + 8, y1 + 100], scale=1.5,thickness=2)
            
        if confi<20:
            mixer.music.stop()
            
    cv2.imshow('frame',frame)
    cv2.waitKey(1)

    if cv2.waitKey(20) & 0xFF==ord('k'):
        break

server.quit()
cap.release()
cv2.destroyAllWindows()
```
